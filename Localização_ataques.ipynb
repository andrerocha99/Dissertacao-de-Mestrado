```python
import matplotlib
import matplotlib.pyplot as plt
import geopandas as gpd
import plotly.express as px
import pandas as pd
```

```python
countries_path = r'C:\Users\35192\Desktop\Tese\Criar Base de Dados\Base_Dados.csv'
```

```python
df = pd.read_csv(countries_path)
df
```

```python
#using geopandas to convert Lon and Lat to points
df_geo = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(
df.lon_d, df.lat_d))    
    
```

```python
df_geo
```

```python
#get built in dataset from geopandas
world_data = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
#plot world map
axis = world_data [world_data.continent == 'Africa'].plot(
    color = 'lightblue', edgecolor = 'black')

df_geo.plot(ax= axis, color = 'red')
plt.title('Golfo da Guiné Ataques')

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10,6)
fig.savefig('matplot.png', dpi = 200)
plt.show()
```
