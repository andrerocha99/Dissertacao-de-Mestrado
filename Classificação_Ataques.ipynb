```python
def select_marker_color(row):
    if row['Classificao_Ataque'] == 'DANIFICADO':
        return 'pink'
    elif row['Classificao_Ataque'] == 'HIJACK'or row['Classificao_Ataque'] == 'HIJACK/ROUBO':
        return 'orange'
    elif row['Classificao_Ataque'] == 'HOMICIDIO' or row['Classificao_Ataque'] == 'HIJACK/HOMICIDIO/ROUBO' or  row['Classificao_Ataque'] == 'HOMICIDIO/RAPTO' or  row['Classificao_Ataque'] == 'HOMICIDIO/ROUBO':
        return 'red'
    elif row['Classificao_Ataque'] == 'SEQUESTRO' or  row['Classificao_Ataque'] == 'HIJACK/ROUBO/SEQUESTRO' or  row['Classificao_Ataque'] == 'ROUBO/SEQUESTRO' or  row['Classificao_Ataque'] == 'RAPTO/ROUBO/SEQUESTRO' or  row['Classificao_Ataque'] == 'HIJACK/SEQUESTRO':
        return 'blue'
    elif row['Classificao_Ataque'] == 'RAPTO' or  row['Classificao_Ataque'] == 'HIJACK/RAPTO' or  row['Classificao_Ataque'] == 'HIJACK/RAPTO/ROUBO' or  row['Classificao_Ataque'] == 'RAPTO/ROUBO':
        return 'black'
    elif row['Classificao_Ataque'] == 'ROUBO':
        return 'gray'
    return 'green' #se a classificao do ataque for no conseguido
```

```python
ataques['color'] = ataques.apply(select_marker_color, axis=1)
```

```python
for _, ataque in ataques.iterrows():
   folium.Marker(
       location= [ataque['lat_d'], ataque['lon_d']],
       tooltip= [ataque['Data'],ataque['Hora_UTC']],
       icon=folium.Icon(color=ataque['color']),
   ).add_to(my_map)

my_map
```
