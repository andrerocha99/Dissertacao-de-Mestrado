```python
import plotly
import plotly.express as px
import pandas as pd
```

```python
countries_path = r'C:\Users\35192\Desktop\Tese\Criar Base de Dados\Base_Dados.csv'
```

```python
df = pd.read_csv(countries_path)
df
```

```python
map_fig = px.scatter_geo(df,
                         locations = 'iso_alpha',
                         projection = 'orthographic',
                         color = 'Ano',
                         opacity = .8,
                         hover_name = 'Ano',
                         hover_data = ['Data', 'Estado_Costeiro']
                        )
```

```python
plotly.offline.plot(map_fig, filename = 'C:\map_exp.html', )
```
