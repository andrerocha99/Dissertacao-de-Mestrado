```python
#Gráfico de linhas para demonstrar a evolução anual do número de ataques de pirataria
import matplotlib.pyplot as plt
import numpy as np

xpoints= np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
ypoints= np.array([43, 45, 39, 56, 61, 52, 27, 30, 51, 44, 65, 60, 63, 0])

plt.plot(xpoints, ypoints, marker = 'o')

plt.xlabel("Ano a considerar")
plt.ylabel("Número de Ataques")
plt.title("Evolução Anual do número de ataques de pirataria")

plt.grid(color= 'black', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
#Gráfico para analisar a relação entre o Estado do Navio e a ocorrência de um ataque

#Linha 1 (azul) representa os ataques a navios que se encontravam fundeados
x1 = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
y1 = np.array([26, 23, 23, 33, 33, 21, 9, 18, 16, 16, 33, 29, 20 , 0])

#Linha 2 (cor de laranja) representa os ataques a navios que se encontravam a navegar
x2 = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
y2 = np.array([10, 19, 15, 22, 27, 29, 17, 9, 30, 28, 30, 27, 38, 0])

#Linha 3 (verde) representa os ataques a navios que se encontravam atracados
x3 = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021])
y3 = np.array([7, 3, 1, 1, 1, 2, 1, 3, 5, 0,2, 4, 5, 0])

plt.ylabel("Número de Ataques")
plt.title("Número de ataques consoante o Estado do Navio ")

plt.plot(x1,y1, color = 'blue', label="Fundeado", marker = 'o')
plt.plot(x2,y2, color = 'orange', label="Navegar", marker = 'o')
plt.plot(x3,y3, color = 'green', label="Atracado", marker = 'o')

plt.legend(loc="upper right")
plt.grid(color= 'black', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
#Pie Chart para comparar a proporção de ataques para cada Estado do Navio
y = np.array([13,370,253])
Estado_Navio = ["Atracado", "Fundeado", "Navegar"]

plt.pie(y,labels = Estado_Navio, autopct='%1.1f%%', shadow=True )
plt.legend(title = "Estados do Navio")
plt.show()


```
